<h4>Contiene información técnica y de mercado de automóviles. El objetivo es predecir el precio del automóvil en función de sus atributos</h4>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
Dato = pd.read_csv('D:/Car details v3.csv')

Datos = pd.DataFrame(Dato)
```

```python
Datos.head(100)
```

```python
# Supongo que 'Datos' es un DataFrame de pandas previamente definido
x = Datos[['year', 'km_driven']]
y = Datos['selling_price']

modelo_lineal = LinearRegression()

# Ajustar el modelo
modelo_lineal.fit(x, y)

beta_0 = modelo_lineal.intercept_
beta_1, beta_2 = modelo_lineal.coef_

print(f"Intercepto (β0): {beta_0}")
print(f"Coeficiente (year, β1): {beta_1}")
print(f"Coeficiente (km, β2): {beta_2}")  # Corrige aquí para imprimir β2

# Realiza la predicción usando los datos originales
prediccion = modelo_lineal.predict(x)

# Visualiza los resultados
plt.scatter(Datos['km_driven'], y, color='blue', label='Datos Reales')  # Corrige la variable en el eje x
plt.plot(Datos['km_driven'], prediccion, color='red', label='Línea de Regresión')  # Corrige aquí también
plt.title('Regresión Lineal: Precio vs km conducidos')  # Mejora el título
plt.xlabel('km conducidos')
plt.ylabel('Precio (USD)')
plt.legend()
plt.show()
```

```python
# Lista de años y kilometrajes correspondientes para predicciones
años = [2007, 2009, 2009, 2015]
km = [100000, 50000, 50012, 15]  
engine = [2500, 1000, 6000, 1800]

# Realizar y mostrar las predicciones para cada año
for i, año in enumerate(años):
    nuevo_x = np.array([[año, km[i]]])
    prediccion = modelo_lineal.predict(nuevo_x)
    print(f"Predicciones para el año {año} con {km[i]} km: {prediccion[0]:.2f} USD")
```

```python
x = Datos[['year', 'km_driven']]  
y = Datos['selling_price'] 
model = LinearRegression()

model.fit(x,y)

predicciones = model.predict(x)

# R2_score: mide que tan cercano es el valor predicho con el real (en un rango de variabilidad de las variables). 
# Valores entre 0 y 1, entre mas cercano a 1 es mejor
r2 = r2_score(y,predicciones) 
print(f"R2 coeficiente de determinacion: {r2}")

# MAE, Error absoluto medio: mide el error promedio
mae = mean_absolute_error(y,predicciones)
print(f"Error absoluto medio (MAE): {mae}")

# MSE, Error cuadratico medio: Evalua el promedio cuadrado de los errores. Entre mas bajo es mejor 
mse = mean_squared_error(y,predicciones)
print(f"Error cuadratico medio (mse): {mse}")

# RMSE, Raiz del error cuadratico medio: Raiz cuadrada de MSE
rmse = np.sqrt(mse)
print(f"error raiz del error cuadratico medio (rmse): {rmse}")

# MAPE, Error absoluto medio porcentual. Entre mas bajo el valor mejor es el modelo
def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
mape1 = mape(y,predicciones)
print(f"Porcentaje de error: {mape1}")
```

<h3>Analisis: Tras la realizacion del modelo y la evaluación los porcentajes fueron los siguientes</h3>

<h4>- R2_score: 0,1748. Este valor indica que aproximadamente el 17.48% de la variación en los datos se puede explicar por el modelo. Un R2 tan bajo sugiere que el modelo no está capturando bien la relación entre las variables, lo que podría indicar que no es un modelo adecuado para los datos.</h4>

<h4>- MAE: 386122,310. Este valor representa el promedio de las diferencias absolutas entre las predicciones y los valores reales. Un MAE de 386122.31 sugiere que, en promedio, el modelo se equivoca en esta cantidad</h4>

<h4>- MSE: 536347916608,51. El MSE penaliza más los errores grandes, ya que los errores se elevan al cuadrado. Este valor es bastante alto, lo que también indica que el modelo tiene errores significativos en algunas predicciones</h4>

<h4>- RMSE: 732357,77. El RMSE se expresa en las mismas unidades que la variable dependiente y se interpreta de manera similar al MAE. Un RMSE de 732357.78 indica un error considerable en las predicciones</h4>

<h4>- MAPE, Porcentaje de error: 88,14%. Un porcentaje de error del 88.15% es bastante alto y sugiere que el modelo tiene un rendimiento deficiente, ya que las predicciones son, en promedio, inexactas en más del 88% en relación con los valores reales</h4>
